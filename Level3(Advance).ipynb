```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the text sentiment dataset
sentiment_df = pd.read_csv(r"D:\codeveda internship\3) Sentiment dataset.csv")

# 2. Clean extra spaces from the text categories
sentiment_df['Sentiment'] = sentiment_df['Sentiment'].str.strip()

# 3. Print the count of each sentiment
print("Sentiment Counts:\n", sentiment_df['Sentiment'].value_counts())

# 4. Draw a simple bar chart to show the counts
plt.figure(figsize=(6, 4))
sns.countplot(x='Sentiment', data=sentiment_df)
plt.title("Text Sentiment Distribution")
plt.show()
```
