```python
#Task 1 Data Cleaning
import pandas as pd
```

```python
df = pd.read_csv(r"D:\codeveda internship\1) iris.csv")
```

```python
df.isnull().sum()
df.dropna(inplace=True)

df.drop_duplicates(inplace=True)

df.columns = df.columns.str.strip().str.lower()

df.to_csv("cleaned_iris_level1.csv", index=False)
```

```python
#Task 3 Data Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cleaned data safely load karo
df = pd.read_csv("cleaned_iris_level1.csv")

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar Plot: LSepal Length By Species
sns.barplot(x=df.columns[-1], y=df.columns[0], data=df)
plt.title("Sepal Length By Species")
plt.show()

# 2. Scatter Plot: Sepal Length vs Sepal Width
sns.scatterplot(x=df.columns[0], y=df.columns[1], hue=df.columns[-1], data=df)
plt.title("Length vs Width")
plt.show()
```
